```python
import pandas as pd
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("../dados/titanic.csv")
df
```

```python
df.columns
```

```python
fig = px.scatter(df, x = "Fare", y = "Ticket",
                 color = "Pclass",
                 size = "SibSp",
                 hover_name="Name"
                )

fig.update_layout(hovermode="x")

fig.update_layout(xaxis = {'title':'Ticket Pago'}, yaxis = {'title':'Tarifa'})

fig.update_xaxes(showspikes=True, spikecolor="green", spikesnap="cursor", spikemode="across", spikethickness=1)
fig.update_yaxes(showspikes=True, spikecolor="Black", spikethickness=1)

fig.show()
```

## Importação de dados financeiros de APIs

Obs.: Para baixar o pandas-datareader, precisamos entrar no prompt de comando do Anaconda e usar os comandos do PIP (pip install pandas-datareader)

```python
import pandas_datareader as wb
import pandas as pd
```

```python
df = wb.DataReader("PETR4.SA", data_source = "yahoo", start = "2018-12-31")
df
```

Podemos selecionar somente uma coluna:

```python
df["Adj Close"]
```

Para importar o fechamento ajustado de várias ações, criamos um df vazio, extraímos o fechamento de cada df importado e adicionamos ao df novo.

```python
tickers = ["PETR4.SA", "VALE3.SA", "^BVSP"]

df_1 = pd.DataFrame()

for i in tickers: # iterações dentro da lista

    df_1[i] = wb.DataReader(i, data_source = "yahoo", start = "2018-12-31")["Adj Close"]

df_1
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
lista = np.array([9, 21, 7, 25, 13, 21, 5, 20, 4, 19])
print("Média: ", lista.mean())
print("Mediana: ", np.median(lista))
print("Range: ", lista.max() - lista.min())
print("Desvio-Padrão: ", lista.std(ddof=1).round(1))
print("Variância: ", lista.var(ddof=1).round(1))
```

```python
media = 3
desvio = 1.2
dist = norm(loc=media, scale=desvio)

print(dist.cdf(3) - dist.cdf(2))
print(1 - dist.cdf(1))
```

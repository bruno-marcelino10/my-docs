# Testes

Os testes de software so processos para verificar se um programa ou aplicativo opera conforme o esperado e atende aos requisitos estabelecidos. Eles so cruciais para garantir a qualidade do software antes de ser lanado para os usurios finais.

Existem vrios nveis de testes de software, que so organizados de acordo com o escopo e os objetivos especficos de verificao. Cada nvel de teste aborda diferentes aspectos e nveis de detalhe do software, contribuindo para a qualidade e confiabilidade do produto final.

No Python, a palavra reservada `assert`  normalmente utilizada para criar testes condicionais, onde caso alguma condio no  satisfeita, retorna um `AssertionError`. 

**Hierarquia de Testes**



### Teste Unitrio


Encontra falhas de funcionamento dentro de uma pequena parte do sistema, **isolando-a** para teste, sem depender do restante do sistema ou quaisquer fatores externos que j estejam implementados.

Concentra-se na verificao individual das partes ou unidades mais bsicas do software, como funes ou mtodos. Geralmente so feitos pelos prprios desenvolvedores, sendo utilizados tambm para o Desenvolvimento Orientado a Testes (TDD).

```python
# funo de adio
def soma(a, b):
    return a + b

# Arquivo de teste: test_soma.py
def test_soma_ok():
  
  # Verificando se a funo soma retorna o resultado esperado
  assert soma(2, 3) == 5  # Esperado: 2 + 3 = 5
  assert soma(-1, 1) == 0  # Esperado: -1 + 1 = 0
  print("Teste Finalizado!")

test_soma_ok()
```

```python
# Arquivo de teste: test_soma.py
def test_soma_error():
  '''Verificando se a funo soma retorna o resultado esperado'''
  assert soma(2, 3) == 4, "Deveria ser 5!" # *No Esperado:* 2 + 3 != 5
  assert soma(-1, 1) == 0  # Esperado: -1 + 1 = 0

test_soma_error()
```

### Teste de Integrao

Encontra falhas em partes integradas de um sistema, que so testadas em **conjunto**. Seu foco principal consiste em encontrar falhas na transmisso de dados entre unidades de cdigo previamente testadas.

```python
# funo de adio
def soma(a, b):
    return a + b

# funo de multiplicao
def multiplicacao(a, b):
    return a * b

# funo que utiliza soma e multiplicao
def operacao_complexa(x, y, z):
    return multiplicacao(soma(x, y), z)

# Arquivo de teste: test_operacao_complexa.py
def test_operacao_complexa():
    '''Verificando se a funo operacao_complexa retorna o resultado esperado'''
    assert operacao_complexa(2, 3, 4) == 20  # Esperado: (2 + 3) * 4 = 20
    assert operacao_complexa(0, 5, 2) == 10  # Esperado: (0 + 5) * 2 = 10
    print("Teste Finalizado!")

test_operacao_complexa()
```

### Teste de Sistema

Encontra falhas no sistema por completo, i.e., verificando se os componentes so compatveis, interagem corretamente e transferem os dados no momento certo. Avalia se  possvel colocar o sistema em produo, sem quaisquer tipos de bugs.

Partindo da documentao, o sistema  testado por completo para verificar se os resultados obtidos esto coerentes com o que era esperado para a ferramenta.

Pode abranger testes de funcionalidade, desempenho, segurana, entre outros.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-10, 10, 100)
sig = 1 / (1 + np.exp(-t))

plt.axhline(y=0, color="black", linestyle="--") # linha tracejada no eixo y = 0
plt.axhline(y=0.5, color="black", linestyle=":") # linha pontilhada no eixo y = 0.5
plt.axhline(y=1.0, color="black", linestyle="--") # linha tracejada no eixo y = 1
plt.axvline(color="grey") # linha no eixo x = 0
plt.axline((0, 0.5), slope=0.25, color="black", linestyle=(0, (5, 5))) # linha reta com inclinação
plt.plot(t, sig, linewidth=2, label=r"$\sigma(t) = \frac{1}{1 + e^{-t}}$") # Acrescenta o plot e códigos do LaTeX na legenda
plt.xlim(-10, 10) 
plt.xlabel("t")
plt.legend(fontsize=14)
plt.show()
```

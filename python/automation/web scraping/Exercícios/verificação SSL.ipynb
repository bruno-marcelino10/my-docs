```python
import requests
from bs4 import BeautifulSoup
```

```python
req = requests.get("https://www.b3.com.br/pt_br/market-data-e-indices/servicos-de-dados/market-data/consultas/mercado-a-vista/participacao-dos-investidores/volume-total-acumulado/")
print(req) # 200 = deu bom
```

```python
html = req.text
soup = BeautifulSoup(html, "html.parser")
```

```python
print(soup.prettify())
```

```python
soup.find_all("iframe")[1]["src"] # retorna a página onde consta a tabela
```

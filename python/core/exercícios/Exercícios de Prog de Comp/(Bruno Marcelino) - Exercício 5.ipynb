````python
def fatorial_1(n, i): 
    
    if i == n:    
        resultado = n
    else:
        resultado = i * fatorial_1(n, i + 1)
    
    return(resultado)

def exercicio_1_1(n, i): 
    """
    n = int
    i = 1
    """
    
    if i == 1 and n > 0:
        resposta = fatorial_1(n, i)
    elif i == 1 and n == 0:
        resposta = 1
    else:
        resposta = None
        
    return(resposta)
````

# Exercício 2

n = número de termos da série harmônica

```
SE n < 1:
    resultado inexistente (o número de termos deve ser um natural que começa a partir do 1)
    
    CASO CONTRÁRIO:
       ENQUANTO n >= 1: começando do último termo, na posição n
           SE n for ímpar:
               somamos 1/n ao valor da série
           CASO CONTRÁRIO (n é par):
               subtraímos 1/n  ao valor da série
       
       passamos para o termo da posição n - 1 e repetimos a operação até somar todos os termos das posições 1 até n. 
```

````python
def exercicio_2(n):
    
    if n < 1:
        resultado = None
    else:
        resultado = 0
        while n >= 1:
            if n % 2 == 1: # Ímpar
                resultado = resultado + 1/n
            else:
                resultado = resultado - 1/n
            n = n - 1
            
    return(resultado)
````
